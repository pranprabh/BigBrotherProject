```python
# This sample data provides insight into the nuances in matching Bigs and Littles.
# There is more to helping the Littles than getting them to answer questions.
# We've shown here that happiness in the state they live in can be a key factor in deciding how to allocate resources
# Now, let's look at how COVID-19 could change how we help our Littles:
```

```python
# If we make the assumption that COVID-19 has left a drastic mark on communities that have been hit hard,
# we can infer that the Littles' responses may indicate lower levels of overall health. This requires more attention.
# Let's say Littles in states where COVID-19 has spread to more that 250,000 individuals causes their average survey
# result to be 3, while all states where COVID-19 has spread to between 50,000 individuals and 250,000 individuals
# causes their average survey result to be 6. Finally, we can say Littles on average score a 7 on the survey for 
# other states.

# Highly infected populations: California, Texas, Florida, New York, Georgia, Illinois
# Moderately infected populations: North Carolina, Arizona, Tennessee, New Jersey, Pennsylvania, Ohio, Lousiana, Alabama, Virginia, South Carolina, Wisconsin, Michigan, Missouri, Massachusetts, Indiana, Maryland, Minnesota, Mississippi, Oklahoma, Iowa, Washington, Arkansas, Utah, Nevada, Kentucky, Colorado, Kansas, Connecticut, Nebraska
# Other infected populations: Idaho, Oregon, New Mexico, South Dakota, North Dakota, Rhode Island, Delaware, Montana, West Virginia, Hawaii, Alaska, New Hampshire, Wyoming, Maine, Vermont

moderately_infected = ['North Carolina', 'Arizona', 'Tennessee', 'New Jersey', 'Pennsylvania', 'Ohio', 'Louisiana', 'Alabama', 'Virginia', 'South Carolina', 'Wisconsin', 'Michigan', 'Missouri', 'Massachusetts', 'Indiana', 'Maryland', 'Minnesota', 'Mississippi', 'Oklahoma', 'Iowa', 'Washington', 'Arkansas', 'Utah', 'Nevada', 'Kentucky', 'Colorado', 'Kansas', 'Connecticut', 'Nebraska']
other_infected = ['Idaho', 'Oregon', 'New Mexico', 'South Dakota', 'North Dakota', 'Rhode Island', 'Delaware', 'Montana', 'West Virginia', 'Hawaii', 'Alaska', 'New Hampshire', 'Wyoming', 'Maine', 'Vermont']
```

```python
state_need_pairs = []

for i in range(len(state_happinesses)):
    if states[i] in highly_infected:
        state_need_pairs.append({"code": states[i], "value": reg.predict([[state_happinesses[i], 3]])})
    elif states[i] in moderately_infected:
        state_need_pairs.append({"code": states[i], "value": reg.predict([[state_happinesses[i], 6]])})
    elif states[i] in other_infected:
        state_need_pairs.append({"code": states[i], "value": reg.predict([[state_happinesses[i], 7]])})
    else:
        print("error")

for i in range(len(state_need_pairs)):
    state_need_pairs[i]["value"] = state_need_pairs[i]["value"][0]

def by(e):
    return e["value"]

state_need_pairs.sort(key=by)

f = open("state_need_pairs.txt", "x")
f.write(str(state_need_pairs))
f.close()
```
